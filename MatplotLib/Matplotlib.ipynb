```python
pd.set_option('display.max_columns', None)
WorldCupMatches = pd.read_csv('WorldCupMatches.csv')
WorldCupPlayers = pd.read_csv('WorldCupPlayers.csv')
WorldCups = pd.read_csv('WorldCups.csv')
```

```python
invalid_years = WorldCupMatches['Year'][WorldCupMatches['Year'].isna()]   #Identificar los valores no vÃ¡lidos
print(invalid_years)
```

```python
WorldCupMatches = WorldCupMatches.dropna(subset=['Year'])   #Eliminar las filas con valores no vÃ¡lidos
```

```python
#Goles Locales anotados en cada mundial
WorldCupMatches['Year'] = WorldCupMatches['Year'].astype(int)   # Convertir los valores de 'Year' a enteros
plt.plot(WorldCupMatches['Year'], WorldCupMatches['Home Team Goals'])
plt.xticks(range(WorldCupMatches['Year'].min(), WorldCupMatches['Year'].max() + 1, 6))   # AquÃ­ estoy configurando el espaciado para que se muestren los aÃ±os cada 6 aÃ±os
plt.xticks(rotation=45)   # Rotar los ticks del eje x en 45 grados
plt.show()
```
