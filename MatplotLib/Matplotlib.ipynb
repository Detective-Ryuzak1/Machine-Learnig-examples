```python
# Gráfico de barras apiladas
import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C', 'D']
valores1 = [20, 35, 30, 35]
valores2 = [25, 32, 34, 20]

plt.bar(categorias, valores1, label='Grupo 1')   # Gráfica los valores de grupo 1
plt.bar(categorias, valores2, bottom=valores1, label='Grupo 2')   # Gráfica los valores de grupo 1

plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de barras apiladas')
plt.legend()   # Coloca una leyenda para identificar los grupos
plt.show()
```

```python
# Gráfico de barras agrupadas
import numpy as np
import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C', 'D']
valores1 = [20, 35, 30, 35]
valores2 = [25, 32, 34, 20]

x = np.arange(len(categorias))
ancho = 0.45

plt.bar(x - ancho/2, valores1, width=ancho, label='Grupo 1')
plt.bar(x + ancho/2, valores2, width=ancho, label='Grupo 2')

plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de barras agrupadas')
plt.xticks(x, categorias)
plt.legend()
plt.show()
```

```python
# Gráfico de contorno
x = np.linspace(-3.0, 3.0, 100)   # Genera un arreglo de 100 números equiespaciados en el intervalo de -3.0 a 3.0
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)   # Crea una cuadrícula de coordenadas bidimensional a partir de dos arreglos unidimensionales x y y
Z = np.sin(np.sqrt(X**2 + Y**2))   # Calcula el seno del cuadrado de la raíz de la suma de los cuadrados de las matrices X y Y. Esta expresión se encuentra comúnmente en contextos de gráficos tridimensionales, donde se quiere calcular un valor para cada punto de una superficie.

plt.contour(X, Y, Z)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de contorno')
plt.show()
```

```python
# Convertir la masa corporal de string a valores númericos
new_weights = []   # Crea una lista para almacenar los nuevos valores de peso

# Itera sobre cada elemento `x` en la columna 'Weight' y aplica la lógica de conversión
for x in fifa.Weight:
    if type(x) == str:   # Si el elemento es una cadena (str), elimina los caracteres 'kg' y convierte a entero
        new_weight = int(x.strip('kg'))

    else:   # Si no es una cadena, mantén el valor sin cambios
        new_weight = x
    new_weights.append(new_weight)   # Agrega el nuevo valor de peso a la lista

# Asigna la lista de nuevos pesos a la columna 'Weight' del DataFrame `fifa`
fifa.Weight = new_weights

print(fifa.Weight[0])
print(type(fifa.Weight[0]))
```

```python
# Clasifica los jugadores por pesos en kg
light = fifa.loc[fifa.Weight < 60].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 60) & (fifa.Weight < 75)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 75) & (fifa.Weight < 85)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >= 85) & (fifa.Weight < 95)].count().iloc[0]
heavy = fifa.loc[fifa.Weight > 95].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['muy delgado', 'delgado', 'intermedio', 'pesado', 'muy pesado']
explode = (0, 0, 0, .2, 1)
plt.style.use('ggplot')   # Cambia los colres usados por default para la gráfica
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)   # Crea una gráfica de pastel desplazando los porcentajes a la orilla y 
plt.title('Masa corporal de los jugadores')
plt.show()
```

```python
unique_nationalities = fifa['Nationality'].unique()   # Obtiene todas las nacionalidades únicas

player_counts = []   # Lista para almacenar el número de jugadores por nacionalidad

# Itera sobre cada nacionalidad y cuenta el número de jugadores
for nationality in unique_nationalities:
    count_players = fifa.loc[fifa['Nationality'] == nationality].count().iloc[0]
    player_counts.append(count_players)

explode = tuple(np.full(160, 2))

# Configura el gráfico de pastel
plt.figure(figsize=(10, 10))
plt.pie(player_counts, labels=unique_nationalities, autopct='%1.1f',pctdistance=0.8, explode=explode)
plt.title('Distribución de jugadores por nacionalidad', fontdict={'fontweight':'bold', 'fontsize':18})
plt.axis('equal')  # Asegura que el gráfico sea un círculo
plt.show()
```

```python
# Gráfico de superficie 3D
from mpl_toolkits.mplot3d import Axes3D  # Para gráficos 3D


fig = plt.figure()   # Crea una nueva figura

ax = fig.add_subplot(111, projection='3d')   # Agregar un subplot en 3D a la figura
x = np.linspace(-5, 5, 100)  # Crea 100 puntos espaciados uniformemente entre -5 y 5 en el eje x
y = np.linspace(-5, 5, 100)  # Crea 100 puntos espaciados uniformemente entre -5 y 5 en el eje y

# Crea una cuadrícula de coordenadas a partir de los datos x e y
X, Y = np.meshgrid(x, y)  # Crea una cuadrícula 2D de coordenadas a partir de x e y

# Calcula los valores de z utilizando la función sin(sqrt(x^2 + y^2))
Z = np.sin(np.sqrt(X**2 + Y**2))  # Calcula el seno de la raíz cuadrada de la suma de los cuadrados de X y Y

ax.plot_surface(X, Y, Z)  # Graficar la superficie 3D utilizando los datos X, Y, Z
plt.title('Gráfico de superficie 3D')
plt.show()
```

```python
# Gráfico de dispersión 3D
fig = plt.figure()   # Crea una nueva figura

# Agrega un subplot en 3D a la figura
ax = fig.add_subplot(111, projection='3d')  # '111' significa que estamos creando una cuadrícula de 1x1 subgráficos y este es el primer (y único) subplot

# Generar datos aleatorios para los ejes x, y, z
x = np.random.standard_normal(100)  # Crea 100 valores aleatorios distribuidos normalmente para el eje x
y = np.random.standard_normal(100)  # Crea 100 valores aleatorios distribuidos normalmente para el eje y
z = np.random.standard_normal(100)  # Crea 100 valores aleatorios distribuidos normalmente para el eje z

# Grafica los puntos en el espacio tridimensional
ax.scatter(x, y, z)  # Grafica los puntos en el espacio tridimensional

plt.title('Gráfico de dispersión 3D')
plt.show()
```

```python
categorias = ['A', 'B', 'C', 'D']  # Etiquetas de las categorías
valores1 = [20, 35, 30, 35]  # Alturas de las barras para el primer conjunto de datos
valores2 = [25, 32, 34, 20]  # Alturas de las barras para el segundo conjunto de datos
x = np.arange(len(categorias))  # Genera una secuencia de números para ubicar las barras en el eje x
fig = plt.figure()   # Crea una nueva figura

# Agrega un subplot en 3D a la figura
ax = fig.add_subplot(111, projection='3d')  # '111' significa que estamos creando una cuadrícula de 1x1 subgráficos y este es el primer (y único) subplot

# Itera sobre las categorías y graficar las barras para cada categoría
for i in range(len(categorias)):
    # Grafica la barra para el primer conjunto de datos
    ax.bar3d(x[i], 0, 0, 0.5, valores1[i], 0.5, color='b', alpha=0.8)  # Definir la posición y dimensiones de la barra, y establecer su color y transparencia
    # Grafica la barra para el segundo conjunto de datos
    ax.bar3d(x[i], 0, 0, 0.5, valores2[i], 0.5, color='r', alpha=0.8)  # Definir la posición y dimensiones de la barra, y establecer su color y transparencia


ax.set_xlabel('Categorías')
ax.set_ylabel('Valores')
ax.set_zlabel('Z')
ax.set_title('Gráfico de barras 3D')

# Ajusta manualmente la posición de la leyenda
max_val = max(max(valores1), max(valores2))
ax.text(x[-1], max_val, 0, "Grupo 1", color='b', fontsize=10, ha='right', va='bottom')
ax.text(x[-1], max_val, 0, "Grupo 2", color='r', fontsize=10, ha='right', va='top')

# Agrega la leyenda dentro de la gráfica
ax.text2D(0.5, 0.95, 'Leyenda', transform=ax.transAxes, ha='center', fontsize=10)
ax.text2D(0.5, 0.9, 'Grupo 1: Azul', transform=ax.transAxes, ha='center', fontsize=8, color='black')
ax.text2D(0.5, 0.85, 'Grupo 2: Rojo', transform=ax.transAxes, ha='center', fontsize=8, color='black')

plt.show()
```
