```python
fifa = pd.read_csv('FIFA19FullData.csv')
fifa.head(5)
```

```python
#2 GrÃ¡ficas (Histograma y Barras)
fig, axs = plt.subplots(2, 1, figsize=(14, 10))   # Crea la figura y los subplots para mostrar 2 grÃ¡ficos

# Histograma para la columna 'Rating' de fifa
axs[0].hist(fifa['Rating'], bins=100, color='r')   #Crea un histograma con 100 intervalos, en color rojo
axs[0].set_title('Histograma de Ratings en FIFA', fontdict={'fontweight':'bold', 'fontsize':18})
axs[0].set_xlabel('Rating de habilidad')
axs[0].set_ylabel('NÃºmero de jugadores')
plt.setp(axs[0].xaxis.get_majorticklabels(), rotation=45)   # Rota las etiquetas del eje x en el primer subplot para mejorar la legibilidad
axs[0].set_xticks(range(int(fifa['Rating'].min()), int(fifa['Rating'].max()) + 1, 2))   # Establecer los ticks del eje x para que estÃ©n separados por intervalos de 2 unidades en la primera grÃ¡fica

# GrÃ¡fica de barras para la columna 'Team Initials' de WorldCupPlayers
axs[1].bar(WorldCupPlayers['Team Initials'].unique(), WorldCupPlayers['Team Initials'].value_counts())   #Crea una grÃ¡fica de barras usando valores sin repetir para los paÃ­ses
axs[1].set_title('GrÃ¡fica de barras de la nacionalidad de los jugadores en Copa del Mundo', fontdict={'fontweight':'bold', 'fontsize':18})
axs[1].set_xlabel('PaÃ­ses participantes')
axs[1].set_ylabel('NÃºmero de jugadores')
plt.setp(axs[1].xaxis.get_majorticklabels(), rotation=90, ha='right')   # Rota etiquetas x 90 grados y coloca el extremo derecho de cada etiqueta para coincir con la posiciÃ³n de las marcas en el eje x 

plt.tight_layout()   # Ajusta el espacio entre los subplots
plt.show()   # Muestra las grÃ¡ficas
```

```python
#GrÃ¡fica de pastel
left = fifa.loc[fifa['Preffered_Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preffered_Foot'] == 'Right'].count().iloc[0]

labels = ['Left', 'Right']
colors = ['#31f700', '#ad00f7']   #Establece los colores en valor Hexadecimal

plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f %%')   #Muestra el grÃ¡fico de pastel con un valor porcentual de 2 decimales 
plt.title('Pie dominante')
plt.show()
```

```python
#Convertir la masa corporal de string a valores nÃºmericos
new_weights = []   # Crea una lista para almacenar los nuevos valores de peso

# Itera sobre cada elemento `x` en la columna 'Weight' y aplica la lÃ³gica de conversiÃ³n
for x in fifa.Weight:
    if type(x) == str:   # Si el elemento es una cadena (str), elimina los caracteres 'kg' y convierte a entero
        new_weight = int(x.strip('kg'))

    else:   # Si no es una cadena, mantÃ©n el valor sin cambios
        new_weight = x
    new_weights.append(new_weight)   # Agrega el nuevo valor de peso a la lista

# Asigna la lista de nuevos pesos a la columna 'Weight' del DataFrame `fifa`
fifa.Weight = new_weights

print(fifa.Weight[0])
print(type(fifa.Weight[0]))
```

```python
#Clasifica los jugadores por pesos en kg
light = fifa.loc[fifa.Weight < 60].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 60) & (fifa.Weight < 75)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 75) & (fifa.Weight < 85)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >= 85) & (fifa.Weight < 95)].count().iloc[0]
heavy = fifa.loc[fifa.Weight > 95].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['muy delgado', 'delgado', 'intermedio', 'pesado', 'muy pesado']
explode = (0, 0, 0, .2, 1)
plt.style.use('ggplot')   #Cambia los colres usados por default para la grÃ¡fica
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)   # Crea una grÃ¡fica de pastel desplazando los porcentajes a la orilla y 
plt.title('Masa corporal de los jugadores')
plt.show()
```

```python
unique_nationalities = fifa['Nationality'].unique()   # Obtiene todas las nacionalidades Ãºnicas

player_counts = []   # Lista para almacenar el nÃºmero de jugadores por nacionalidad

# Itera sobre cada nacionalidad y cuenta el nÃºmero de jugadores
for nationality in unique_nationalities:
    count_players = fifa.loc[fifa['Nationality'] == nationality].count().iloc[0]
    player_counts.append(count_players)

explode = tuple(np.full(160, 2))

# Configura el grÃ¡fico de pastel
plt.figure(figsize=(10, 10))
plt.pie(player_counts, labels=unique_nationalities, autopct='%1.1f',pctdistance=0.8, explode=explode)
plt.title('DistribuciÃ³n de jugadores por nacionalidad', fontdict={'fontweight':'bold', 'fontsize':18})
plt.axis('equal')  # Asegura que el grÃ¡fico sea un cÃ­rculo
plt.show()
```

```python
plt.style.use('bmh')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Rating']
madrid = fifa.loc[fifa.Club =='Real Madrid']['Rating']

labels = ['FC Barcelona', 'Real Madrid']
boxes = plt.boxplot([barcelona, madrid], labels=labels, patch_artist=True, medianprops={'linewidth':2})
for box in boxes['boxes']:
        box.set(color='#000000', linewidth=2)   #Cambia el color de la orilla
        box.set(facecolor='#ffffff')   #Cambia el color de rellenado

plt.title('Comparativa de planteles')
plt.ylabel('Nivel de juego')
plt.show()   
# La linea roja representa la media
```
