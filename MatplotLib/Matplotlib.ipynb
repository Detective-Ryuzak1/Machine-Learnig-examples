```python
pd.set_option('display.max_columns', None)
WorldCupMatches = pd.read_csv('WorldCupMatches.csv')
WorldCupPlayers = pd.read_csv('WorldCupPlayers.csv')
WorldCups = pd.read_csv('WorldCups.csv')
```

```python
invalid_years = WorldCupMatches['Year'][WorldCupMatches['Year'].isna()]   #Identificar los valores no vÃ¡lidos
print(invalid_years)
```

```python
WorldCupMatches = WorldCupMatches.dropna(subset=['Year'])   #Eliminar las filas con valores no vÃ¡lidos
```

```python
#Goles Locales anotados en cada mundial
WorldCupMatches['Year'] = WorldCupMatches['Year'].astype(int)   # Convertir los valores de 'Year' a enteros
plt.plot(WorldCupMatches['Year'], WorldCupMatches['Home Team Goals'])   #GrÃ¡fica dos columnas
#plt.xticks(range(WorldCups['Year'].min(), WorldCups['Year'].max() + 1, 6))   # AquÃ­ estoy configurando el espaciado para que se muestren los aÃ±os cada 6 aÃ±os
plt.xticks(WorldCups['Year'][::1])   # AquÃ­ estoy configurando el espaciado para que se muestren los aÃ±os iterando cada valor
plt.xticks(rotation=45)   # Rotar los ticks del eje x en 45 grados
plt.show()
```

```python
#Goles anotados en cada mundial
WorldCups['Year'] = WorldCups['Year'].astype(int)   #Convierte la columna 'Year' a valores enteros
plt.figure(figsize=(8,5))   #Establece la grÃ¡fica con una dimensiÃ³n de 8 pulgadas de ancho por 5 de alto
plt.title('Goles anotados durante cada Mundial', fontdict={'fontweight':'bold', 'fontsize':18})   #Pone in tÃ­tulo a la grÃ¡fica
plt.plot(WorldCups['Year'], WorldCups['GoalsScored'], 'b.-', label='Goles Anotados')   #GrÃ¡fica dos columnas
plt.plot(WorldCups['Year'], WorldCups['MatchesPlayed'], label='Partidos Jugados')
plt.xlabel('AÃ±os')
plt.ylabel('Cantidad')
plt.legend()
plt.xticks(WorldCups['Year'][::1])   # AquÃ­ estoy configurando el espaciado para que se muestren los aÃ±os iterando cada valor
plt.xticks(rotation=45)   # Rotar los ticks del eje x en 45 grados
plt.savefig('goles.png', dpi=300)   #Guarda la grÃ¡fica como imagen
plt.show()
```
