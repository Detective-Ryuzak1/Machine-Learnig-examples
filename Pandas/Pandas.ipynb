```python
import pandas as pd
```

```python
# Crea un dataframe a partir del dataset Dragon_Ball_Data_Set.csv, y configura la columna characters como el índice
df = pd.read_csv('Dragon_Ball_Data_Set.csv', index_col='Character')
```

```python
# Muestra las primeras cinco filas del set de datos
df.head()
```

```python
# Muestra información general del set de datos
df.info()
```

```python
# Muestra el número de filas y columnas del set de datos
df.shape
```

```python
# Muestra cuantas veces aparece Goku en el set de datos
df.loc["Goku"]
```

```python
# Muestra las columnas Power_Level y Saga_or_Movie de los personajes Goku y Vegeta
df.loc[["Goku", "Vegeta"], ["Power_Level", "Saga_or_Movie"]]
```

```python
# Convierte la columna 'Power_Level' a formato numérico y le quita las comas
df['Power_Level'] = pd.to_numeric(df['Power_Level'].str.replace(',', ''), errors='coerce')
pd.set_option('display.float_format', '{:.1f}'.format)
df.loc["Goku"]
```

```python
# Configurar IPython para mostrar números en formato decimal, haciendo uso de un solo decimal
%precision %.1f
```

```python
# Muestra cual es el valor más alto en la columna Power_Level
max_power_level = df['Power_Level'].max()
max_power_level
```

```python
# Organiza el nivel de pelea de los personajes de abajo hacia arriba
df.sort_values(by=['Power_Level'], ascending=False)
```

```python
# Muestra personajes con un nivel de pelea de 4
df.loc[df['Power_Level'] == 4]
```

```python
# Crea un análisis estadístico para la columna Power_Level
df['Power_Level'].describe()
```

```python
# Muestra el valor más repetido de la columna 'Power_Level'
df['Power_Level'].mode()
```

```python
# Muestra el número de veces que se repite un valor en la columna 'Power_Level'
df['Power_Level'].value_counts().head(20)
```

```python
# Muestra personajes con un poder de 10000000000
df.loc[df['Power_Level'] == 10000000000].head(10)
```

```python
# Muestra personajes con un poder de 5 y los organiza tomando en cuenta la columna 'personaje'
df.loc[df['Power_Level'] == 5].sort_values(by='Character')
```

```python
# Filtra los personajes que que tengan un poder de 5 y pertenazcan a la Serie de Dragon Ball
df.loc[(df['Power_Level'] == 5) & (df['Dragon_Ball_Series'] == 'Dragon Ball')]
```

```python
#Crea una nueva columna con el doble del valor poder
df['Kaio-ken'] = df['Power_Level'] * 2
df.loc["Goku"]
```

```python
# Muestra el personaje con el valor Kaio-ken mas alto
df.loc[df['Kaio-ken'] == df['Kaio-ken'].max()]
```

```python
# Muestra el número de personajes que tienen un poder de pelea de 130
df.loc[df['Power_Level'] == 130]
```

```python
forma1 = df.loc[df['Power_Level'] == 130].shape
forma2 = df.query("Power_Level == 130").shape
print("el número de personajes con 130 es:", forma1[0])
print("el número de personajes con 130 es:", forma2[0])
```

```python
#Organiza a Vegeta por nivel de pelea
df.query("Character == 'Vegeta'").sort_values(by="Power_Level")
```

```python
# calcula la media (o promedio) de los valores en la columna 'Power_Level'
mean_power = df['Power_Level'].mean()
mean_power
```

```python
# Muestra personajes que tengan un nivel de pelea superior a la media (el simbolo '`' se usa para encerrar variables con espaciado y el simbolo '@' nos ayuda a referenciar una variable externa)
df.query("`Power_Level` > @mean_power")
```
